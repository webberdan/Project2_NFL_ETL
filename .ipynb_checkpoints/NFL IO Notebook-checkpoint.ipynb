```python
# Create Classes
# ----------------------------------

# Sets an object to utilize the default declarative base in SQL Alchemy
Base = declarative_base()


# Creates Classes which will serve as the anchor points for our Tables
class Stadiums(Base):
    __tablename__ = 'team'
    StadiumID = Column(Integer, primary_key=True)
    Home_Team_Name = Column(String(255))
    City = Column(String(255))
    State = Column(String(255))
    Country = Column(String(255))
    Capacity = Column(String(255))
    PlayingSurface = Column(String(255))
    GeoLat = Column(Float)
    GeoLong = Column(Float)
    Type = Column(String(255))


class Teams(Base):
    __tablename__ = 'stadium'
    Home_team = Column(String(255), primary_key=True)
    Name = Column(String(255))
    Conference = Column(String(255))
    Division = Column(String(255))
    FullName = Column(String(255))
    StadiumID = Column(Integer)
    HeadCoach2022 = Column(String(255))
    OffensiveCoordinator2022 = Column(String(255))
    SpecialTeamsCoach2022 = Column(String(255))
    OffensiveScheme2022 = Column(String(255))
    DefensiveScheme2022 = Column(String(255))
    UpcomingFanDuelSalary = Column(String(255))
    UpcomingDraftKngsSalary = Column(String(255))
    UpcomingYahooSalary = Column(String(255))
    PrimaryColor = Column(String(255))
    SecondaryColor = Column(String(255))
    TertiaryColor = Column(String(255))
    QuaternaryColor = Column(String(255))
    GlobalTeamID = Column(String(255))
    DraftKingsName = Column(String(255))
    FanDuelName = Column(String(255))
    FanDuelPlayerID = Column(String(255))
    FantasyDraftname = Column(String(255))
    FantasyDraftPlayerID = Column(String(255))
    YahooName = Column(String(255))
    YahooPlayerID = Column(String(255))

    
class Games(Base):
    __tablename__ = 'game'
    game_id = Column(String(255), primary_key=True)
    season = Column(String(255))
    game_type = Column(String(255))
    week = Column(String(255))
    gameday = Column(String(255))
    weekday = Column(String(255))
    gametime = Column(String(255))
    away_team = Column(String(255))
    away_score = Column(String(255))
    home_team = Column(String(255))
    home_score = Column(String(255))
    location = Column(String(255))
    result = Column(String(255))
    total = Column(String(255))
    overtime = Column(String(255))
    div_game = Column(String(255))
    temp = Column(String(255))
    wind = Column(String(255))
    away_qb_id = Column(String(255))
    home_qb_id = Column(String(255))
    away_qb_name = Column(String(255))
    home_qb_name = Column(String(255))
    away_coach = Column(String(255))
    home_coach = Column(String(255))
    referee = Column(String(255))
```
